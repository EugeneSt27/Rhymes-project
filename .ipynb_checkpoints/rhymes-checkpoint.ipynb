## Корреляция

```python
def spearman_test(data_1, data_2):
    correlation = spearmanr(data_1, data_2)
    return '{0:.2f}\n{1:.2f}'.format(*correlation)
```

```python
correlation = ds_new[range_scale].corr(method = 'spearman')
```

```python
text = [[spearman_test(ds_new[i], ds_new[j]) for i in range_scale] for j in range_scale]
plt.figure(figsize=(15, 6))
heatmap = sns.heatmap(correlation, vmin=-1, vmax=1,  annot=text, fmt='', cmap='BuPu'
)

heatmap.set_title('Корреляционная матрица для количественных признаков', fontdict={'fontsize':20}, pad=12)
heatmap.set_xticklabels(range_scale, rotation = 50, ha='right')
heatmap.set_yticklabels(range_scale, rotation = 30, ha ='center')
plt.show()
```

# Кластеризация

### Декодирование

```python
ds_cluster_0 = ds_new[['Spotify listeners_log', 'Yandex listeners_log']]
```

### Масштабирование

```python
columns = ['Spotify listeners_log', 'Yandex listeners_log']
          # 'All tracks_log']
```

```python
scaler = MinMaxScaler()
ds_cluster_l = scaler.fit_transform(ds_cluster_0) # get dummies
ds_cluster_n = pd.DataFrame(ds_cluster_l, columns=columns)
ds_cluster_n
```

```python
 fig, (ax1) = plt.subplots(ncols=1, figsize=(10, 8))
ax1.set_title('Scaler')
for col in columns:
    sns.kdeplot(ds_cluster_n[col], ax=ax1, label=col)
ax1.legend()
```

```python
 ds_cluster_0[columns] = ds_cluster_n[columns] # get dummies
```

```python
X_numeric = ds_cluster_n # без жанров
```

## K-means

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 42)
    kmeans.fit(X_numeric)
    wcss.append(kmeans.inertia_)
plt.figure(figsize = (10,6))
plt.plot(range(1, 11), wcss)
plt.title('Выбор количества кластеров методом локтя')
plt.xlabel('Количество кластеров')
plt.ylabel('WCSS')
```

```python
model = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 42)
y_pred = model.fit_predict(X_numeric)
```

```python
ds_cluster_n['cluster_kmean'] = y_pred
```

```python
ds_cluster_n[columns] = scaler.inverse_transform(ds_cluster_l)
ds_cluster_n[columns]
```

```python
ds_new['cluster_kmean'] = y_pred
```

```python
result = {}
groups = [ds_new[ds_new['cluster_kmean'] == i]['Revenue mean'].values for i in ds_new['cluster_kmean'].unique() ]
if any([kstest(i, 'norm').pvalue<0.05 for i in groups]):
    if len(groups) >2:
        result['cluster_kmean'] = [kruskal(*groups), levene(*groups, center='median'), levene(*groups, center='mean')]
    else:
        result['cluster_kmean'] = [mannwhitneyu(*groups), levene(*groups, center='median'), levene(*groups, center='mean')]
else:
     result['cluster_kmean'] = [ttest_ind(*groups), levene(*groups, center='median'), levene(*groups, center='mean')]
result
```

# Функции расчета метрик

```python
def scores(reg, X_train, y_train, X_test, y_test):
    pred_train = reg.predict(X_train)
    pred_test = reg.predict(X_test)
    metrics = pd.DataFrame(
        [[mean_absolute_error(y_train, pred_train),
          mean_absolute_error(y_test, pred_test)],
         [mean_absolute_percentage_error(y_train, pred_train),
          mean_absolute_percentage_error(y_test, pred_test)],
          [np.sqrt(mean_squared_error(y_train, pred_train)),
          np.sqrt(mean_squared_error(y_test, pred_test))],
         [r2_score(y_train, pred_train),
          r2_score(y_test, pred_test)]],
        columns=['Train', 'Test'],
        index=['MAE', 'MAPE', 'RMSE', 'R2'])
    return metrics
```

```python
def innate_scores(reg, X_train, y_train, X_test, y_test):
    pred_train = reg.predict(X_train)
    pred_test = reg.predict(X_test)
    metrics_inn = pd.DataFrame(
        [[mean_absolute_error(np.exp(y_train), np.exp(pred_train)),
          mean_absolute_error(np.exp(y_test), np.exp(pred_test))],
         [mean_absolute_percentage_error(np.exp(y_train), np.exp(pred_train)),
          mean_absolute_percentage_error(np.exp(y_test), np.exp(pred_test))],
          [np.sqrt(mean_squared_error(np.exp(y_train), np.exp(pred_train))),
          np.sqrt(mean_squared_error(np.exp(y_test), np.exp(pred_test)))],
         [r2_score(np.exp(y_train), np.exp(pred_train)),
          r2_score(np.exp(y_test), np.exp(pred_test))]],
        columns=['Train', 'Test'],
        index=['MAE', 'MAPE', 'RMSE', 'R2'])
    return metrics_inn
```

```python
def reverse_scores(reg, X_train, y_train, X_test, y_test,  scaler_y):
    pred_train = reg.predict(X_train)
    pred_test = reg.predict(X_test)
    rev_y_train = scaler_y.inverse_transform(np.exp(y_train))
    rev_y_test = scaler_y.inverse_transform(np.exp(y_test))
    rev_pred_train = scaler_y.inverse_transform(np.exp(pred_train))
    rev_pred_test = scaler_y.inverse_transform(np.exp(pred_test))
    metrics_rev = pd.DataFrame(
        [[mean_absolute_error(rev_y_train, rev_pred_train),
          mean_absolute_error(rev_y_test, rev_pred_test)],
         [mean_absolute_percentage_error(rev_y_train, rev_pred_train),
          mean_absolute_percentage_error(rev_y_test, rev_pred_test)],
          [np.sqrt(mean_squared_error(rev_y_train, rev_pred_train)),
          np.sqrt(mean_squared_error(rev_y_test, rev_pred_test))],
         [r2_score(rev_y_train, rev_pred_train),
          r2_score(rev_y_test, rev_pred_test)]],
        columns=['Train', 'Test'],
        index=['MAE', 'MAPE', 'RMSE', 'R2'])
    return metrics_rev
```

```python
def reverse__tree_scores(reg, X_train, y_train, X_test, y_test,  scaler_y):
    pred_train = reg.predict(X_train).reshape(-1,1)
    pred_test = reg.predict(X_test).reshape(-1,1)
    rev_y_train = scaler_y.inverse_transform(np.exp(y_train))
    rev_y_test = scaler_y.inverse_transform(np.exp(y_test))
    rev_pred_train = scaler_y.inverse_transform(np.exp(pred_train))
    rev_pred_test = scaler_y.inverse_transform(np.exp(pred_test))
    metrics_rev = pd.DataFrame(
        [[mean_absolute_error(rev_y_train, rev_pred_train),
          mean_absolute_error(rev_y_test, rev_pred_test)],
         [mean_absolute_percentage_error(rev_y_train, rev_pred_train),
          mean_absolute_percentage_error(rev_y_test, rev_pred_test)],
          [np.sqrt(mean_squared_error(rev_y_train, rev_pred_train)),
          np.sqrt(mean_squared_error(rev_y_test, rev_pred_test))],
         [r2_score(rev_y_train, rev_pred_train),
          r2_score(rev_y_test, rev_pred_test)]],
        columns=['Train', 'Test'],
        index=['MAE', 'MAPE', 'RMSE', 'R2'])
    return metrics_rev
```

```python
def compare_table(orig, log, log_scaled):

        columns = pd.MultiIndex.from_product([['Train',  'Test'], ['orig', 'log', 'log_scaled']],
                                     names=['subject', 'type'])
        compare_table = pd.DataFrame(columns = columns)
        lr_compare_0 = pd.DataFrame()
        lr_compare_1 = pd.DataFrame()
        lr_compare_0['orig'] = orig['Train']
        lr_compare_0['log'] = log['Train']
        lr_compare_0['log_scaled'] = log_scaled['Train']
        lr_compare_1['orig'] = orig['Test']
        lr_compare_1['log'] = log['Test']
        lr_compare_1['log_scaled'] = log_scaled['Test']
        compare_table['Train'] = lr_compare_0
        compare_table['Test'] = lr_compare_1

        return compare_table
```

```python
def compare_table_cl(orig, log):

        columns = pd.MultiIndex.from_product([['Train',  'Test'], ['orig', 'log']],
                                     names=['subject', 'type'])
        compare_table = pd.DataFrame(columns = columns)
        lr_compare_0 = pd.DataFrame()
        lr_compare_1 = pd.DataFrame()
        lr_compare_0['orig'] = orig['Train']
        lr_compare_0['log'] = log['Train']
        lr_compare_1['orig'] = orig['Test']
        lr_compare_1['log'] = log['Test']
        compare_table['Train'] = lr_compare_0
        compare_table['Test'] = lr_compare_1
        return compare_table
```

```python
def advance_assessment(ds, y_test, y_pred):
  artist_test = ds.loc[y_test.index, ['Artist', 'Month', 'Royalty rate', 'Revenue mean','Advance amount', 'Revenue adv']]
  artist_test['Pred_revenue'] = y_pred
  artist_test.reset_index(inplace = True)
  artist_test = artist_test[artist_test['Advance amount']>0]
  artist_test = artist_test[artist_test['Month']>6]
  artist_test['Advance_pred'] = artist_test['Pred_revenue']*artist_test['Royalty rate']*6
  artist_test['6_months_revenue_real'] = artist_test['Revenue adv']*artist_test['Royalty rate']
  artist_test['paid_off_real'] =  artist_test['6_months_revenue_real'] - artist_test['Advance amount']
  artist_test['paid_off_model'] =  artist_test['6_months_revenue_real'] - artist_test['Advance_pred']
  real_test = artist_test[artist_test['paid_off_real']<0]['paid_off_real'].sum()
  pred_test = artist_test[artist_test['paid_off_model']<0]['paid_off_model'].sum()
  return round(real_test,2), round(pred_test,2)
```
